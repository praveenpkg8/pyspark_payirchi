```plaintext
rdd = sc.textFile(txt_file)
rdd.collect()
```

Example for using Maps in PySpark

```plaintext
rdd1 = rdd.map(lambda x: x.split(' '))
rdd1.collect()
```

```plaintext
def modify_data(line):
  line_list = line.split(' ')
  line_count_list = []
  for _line in line_list:
    line_count_list.append(len(_line))
  return line_count_list

funcMapRdd = rdd.map(modify_data)
funcMapRdd.collect()
```

Example for FlatMap in PySpark

```plaintext
flatMappedRdd = rdd.flatMap(modify_data)
flatMappedRdd.collect()
```

Filter in PySpark

```plaintext
def check_sum_even(list_):
  '''check sum of list is even'''
  if sum(list_) % 2 == 0:
    return True
  else:
    return False

filterMapRdd = funcMapRdd.filter(check_sum_even)
filterMapRdd.collect()
```

groupByKey and mapValues

```plaintext
grpByRdd = rdd.flatMap(lambda x: x.strip().split(' ')).map(lambda x: (len(x), x))
grpByRdd.groupByKey().mapValues(set).collect()
```

ReuduceByKey

```plaintext
reduceRdd = rdd.flatMap(lambda x: x.strip().split(' ')).map(lambda x: (x, 1))
reduceRdd.reduceByKey(lambda x, y: x + y).collect()
```

Averaging to have practice on map, reduceByKey

```plaintext
monthRdd = sc.textFile(csv_file)
monthRdd.collect()
```

```plaintext
monthAvgRdd = monthRdd.map(lambda x: x.split(',')).map(lambda x: [x[0], x[1], float(x[2])])
monthAvgRdd.collect()
```

```plaintext
reducedMonthRdd = monthAvgRdd.map(lambda x: (x[0], (x[2], 1 )))
avgMonthRdd = reducedMonthRdd.reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1])).map(lambda x: (x[0], x[1][0]/x[1][1]))
avgMonthRdd.collect()
```
