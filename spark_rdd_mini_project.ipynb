```python
student_data_csv = ""
```

```python
rddStudent = sc.textFile(student_data_csv)
header = rddStudent.first()
rddStudent = rddStudent.filter(lambda x: x != header) # Trying to remove header from the file
```

```python
rddStudentList = rddStudent.map(lambda x: x.split(','))
rddStudentList.map(lambda x: ("totalStudent", 1)).reduceByKey(lambda x, y: x + y).collect() # Total Student Count
```

```python
rddStudentList.filter(lambda x: x[1].lower() == 'female')\
                                      .map(lambda x: ("femaleTotalMarkStudent", int(x[5])))\
                                      .reduceByKey(lambda x, y: x + y).collect() # Total Female Student Mark
```

```python
rddStudentList.filter(lambda x: x[1].lower() == 'male')\
                                      .map(lambda x: ("maleTotalMarkStudent", int(x[5])))\
                                      .reduceByKey(lambda x, y: x + y).collect() # Total Male Student Mark
```

```python
rddStudentList.filter(lambda x: int(x[5]) > 50)\
                                      .map(lambda x: ("TotalPassedStudent", 1))\
                                      .reduceByKey(lambda x, y: x + y).collect() # Total Passed Student 
```

```python
rddStudentList.map(lambda x: (x[3], 1)).groupByKey()\
                                       .mapValues(list)\
                                       .map(lambda x: (x[0], sum(x[1]))).collect() # Subject Total Mark
```

```python
rddStudentList.map(lambda x: (x[3], int(x[5]))).groupByKey()\
                                       .mapValues(list)\
                                       .map(lambda x: (x[0], sum(x[1]))).collect() # Subject Total Mark
```

```python
rddStudentList.map(lambda x: (x[3], int(x[5]))).groupByKey()\
                                       .mapValues(list)\
                                       .map(lambda x: (x[0], sum(x[1])/len(x[1]))).collect() # Subject Avergae Mark
```

```python
rddStudentList.map(lambda x: (x[3], int(x[5]))).reduceByKey(lambda x, y: x if x < y else y).collect() # Mininum Mark for each subject
```

```python
rddStudentList.map(lambda x: (x[3], int(x[5]))).reduceByKey(lambda x, y: x if x > y else y).collect() # Maximum Mark for each subject
```

```python
rddStudentList.map(lambda x: (x[1], (int(x[0]), 1)))\
              .reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))\
              .map(lambda x: (x[0], x[1][0]/x[1][1])).collect() # Average Male anad Female Age
```
